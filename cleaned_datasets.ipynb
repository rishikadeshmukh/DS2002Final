```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

globalTrend = pd.read_csv('/content/Mental health Depression disorder Data.csv')
techUsage = pd.read_csv('/content/mental_health_and_technology_usage_2024.csv')

#drop duplicates
global_trends = globalTrend.drop_duplicates()
tech_usage = techUsage.drop_duplicates()

#fill missing values (example)
globalTrend.fillna(0, inplace=True)

#remove userID column since it is unnecessary
techUsage = techUsage.drop(columns=['User_ID'])

#unnecessary data values in Year variable in globalTrend df
remove = ['10000 BCE', '9000 BCE', '8000 BCE', '7000 BCE', '6000 BCE', '5000 BCE',
                     '4000 BCE', '3000 BCE', '2000 BCE', '1000 BCE', '0', '100', '200', '300',
                     '400', '500', '600', '700', '800', '900', '1000', '1100', '1200', '1300',
                     '1400', '1500', '1600', '1700', '1710', '1720', '1730', '1740', '1750',
                     '1760', '1770', '1780', '1790', '1', 'Year']
# remove the specified columns
globalTrend= globalTrend[~globalTrend['Year'].isin(remove)]

globalTrend.head()
```
